```python
from bs4 import BeautifulSoup
import requests
```

```python
def request(url):
    """
    Accepts the  URL.
    """
    hdr = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=hdr)
    assert response.status_code == 200, "HTML status code isn't 200 for {}.".format(url)

    return BeautifulSoup(response.text, "lxml")
```

```python
import os
import sys
import time
import numpy as np
from selenium import webdriver

chromedriver = "~/Downloads/chromedriver"  # path to the chromedriver executable
chromedriver = os.path.expanduser(chromedriver)
print('chromedriver path: {}'.format(chromedriver))
sys.path.append(chromedriver)


#url = 'http://www.foxnews.com/category/shows/hannity/transcript.html'


landing_urls = {'Hannity':'http://www.foxnews.com/category/shows/hannity/transcript.html',
                # 'The Five':'http://www.foxnews.com/category/shows/the-five/transcript.html',
                'Ingraham': 'http://www.foxnews.com/category/shows/ingraham-angle/transcript.html'    
                }

#html = driver.page_source

shows = {}
for show, url in landing_urls.items():
    print('Show: ', show)
    print('URL: ', url)


    driver = webdriver.Chrome(chromedriver)
    driver.get(url)
    
    for i in range(11):
        print(i)
        driver.find_element_by_css_selector('.button.load-more.js-load-more').click()
        r = 0.5 * np.random.randn(1) + 5
        time.sleep(r)


    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")

    link_list = []
    for i, class_ in enumerate(soup.find_all('h2',class_="title")):
        link_list.append((class_.find('a').get_text(), class_.find('a')['href']))
        
    shows[show] = {'urls': link_list,
                   'landing_url': url}
```

```python
import pickle
with open('fox_news_shows.pkl', 'wb') as picklefile:
    pickle.dump(shows, picklefile)
```
